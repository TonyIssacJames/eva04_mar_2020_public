```
import matplotlib.pyplot as plt
x = range(1, 20)
```

```
plt.plot(x, [xi*1 for xi in x])
plt.plot(x, [xi*2 for xi in x])
plt.plot(x, [xi*3 for xi in x])

plt.plot(x, [xi*xi for xi in x])
plt.plot(x, [xi*4 for xi in x])
plt.plot(x, [xi*5 for xi in x])

plt.grid()
plt.show()
```

```
fig=plt.figure(figsize=(15,10))
ax=fig.add_subplot(111)


plt.plot(x, [xi*1 for xi in x], label="Model1")
plt.plot(x, [xi*2 for xi in x], label="Model2")
plt.plot(x, [xi*3 for xi in x], label="Model3")

plt.plot(x, [xi*xi for xi in x], label="Model4")
plt.plot(x, [xi*4 for xi in x], label="Model5")
plt.plot(x, [xi*5 for xi in x], label="Model6")

plt.grid()
plt.legend(loc=2)
plt.show()
```
